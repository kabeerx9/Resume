```python
import networkx as nx
import matplotlib.pyplot as plt


```

```python
# Create Karate Club network
G = nx.karate_club_graph()

# Draw network
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Compute degree centrality
degree_centrality = nx.degree_centrality(G)
print("Degree centrality:")
print(degree_centrality)
```

```python
# Compute closeness centrality
closeness_centrality = nx.closeness_centrality(G)
print("\nCloseness centrality:")
print(closeness_centrality)
```

```python
# Compute betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("\nBetweenness centrality:")
print(betweenness_centrality)
```

d(j,i) : sum of shortest path of all the nodes to i ,

```python
# Compute proximity prestige
proximity_prestige = {}
for node in G.nodes:
    distance_sum = sum(nx.shortest_path_length(G, node).values())
    lengths = nx.shortest_path_length(G, source=1)
    reachable_nodes = len(lengths)

    proximity_prestige[node] = distance_sum/reachable_nodes

print("\nProximity prestige:")
print(proximity_prestige)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

#in case of directed graph , u can create a graph like this :

# G = nx.DiGraph()
# G.add_edges_from([(0, 1), (0, 2), (1, 3), (2, 3), (3, 4), (4, 5), (4, 6), (5, 6), (6, 7), (7, 8), (7, 9)])


G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 3), (2, 3), (3, 4), (4, 5), (4, 6), (5, 6), (6, 7), (7, 8), (7, 9)])

nx.draw(G, with_labels=True)
plt.show()

# in case u want to do this on directed graphs use successors instead of neighbours , and also perform the null check in case there are no successors.

def random_walk(G, start_node, target_node):
    current_node = start_node
    steps = 0
    while current_node != target_node:
        neighbors = list(G.neighbors(current_node))
        current_node = random.choice(neighbors)
        steps += 1
    return steps

start_node = 0
target_node = 9

hitting_time = random_walk(G, start_node, target_node)
print(f'Hitting time: {hitting_time}')

commute_time = hitting_time + random_walk(G, target_node, start_node)
print(f'Commute time: {commute_time}')
```

# ASSIGNMENT 3 PAGE RANK 

```python
import numpy as np

def page_rank(adjacency_matrix, damping_factor=0.85, convergence_threshold=0.0001):
    # Create the transition probability matrix
    row_sum = np.sum(adjacency_matrix, axis=1)
    transition_matrix = adjacency_matrix / row_sum[:, np.newaxis]
    
    # Initialize the PageRank vector
    n = adjacency_matrix.shape[0]
    page_rank_vector = np.ones(n) / n
    
    # Perform the PageRank iterations
    while True:
        new_page_rank_vector = ((1 - damping_factor) / n) + (damping_factor * transition_matrix.T @ page_rank_vector)
        if np.max(np.abs(new_page_rank_vector - page_rank_vector)) < convergence_threshold:
            break
        page_rank_vector = new_page_rank_vector
    
    return page_rank_vector
# Example adjacency matrix
adjacency_matrix = np.array([[0, 0, 1, 0],
                             [1, 0, 0, 0],
                             [0, 1, 0, 1],
                             [0, 1, 0, 0]])
# Example usage
page_rank_vector = page_rank(adjacency_matrix, damping_factor=0.85, convergence_threshold=0.0001)
print(page_rank_vector)
```
